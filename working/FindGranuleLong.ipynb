```python
import requests
from pprint import pprint
```

```python
CMR_OPS = 'https://cmr.earthdata.nasa.gov/search'
url = f'{CMR_OPS}/{"collections"}'
# url = f'{CMR_OPS}/{"granules"}'
```

```python
#cloud hosted and has files/granules in it
response = requests.get(url,
                        params={
                            'cloud_hosted': 'True',
                            'has_granules': 'True',
                        },
                        headers={
                            'Accept': 'application/json',
                        }
                       )
response
```

```python
# 200 = success
response.status_code
```

```python
#help(response)
```

```python
for k, v in response.headers.items():
    print(f'{k}: {v}')
```

```python
response.headers['CMR-Hits']
```

```python
#990 datasets
# pare down to PODAAC only
provider = 'POCLOUD'
response = requests.get(url,
                        params={
                            'cloud_hosted': 'True',
                            'has_granules': 'True',
                            'provider': provider,
                        },
                        headers={
                            'Accept': 'application/json'
                        }
                       )
```

```python
response.headers['CMR-Hits']
```

```python
pprint(response.json()['feed']['entry'][0])
# pretty print first
```

```python
collections = response.json()['feed']['entry']
```

```python
len(collections)
```

```python
for collection in collections:
    print(f'{collection["archive_center"]} {collection["dataset_id"]} {collection["id"]}')
    
```

Now that we know what the collections are, let's search through granules

```python
url = f'{CMR_OPS}/{"granules"}'
```

```python
response = requests.get(url, 
                        params={
                            'concept_id': 'C2036878004-POCLOUD',
                            'temporal': '2020-10-17T00:00:00Z,2020-10-18T23:59:59Z',
                            'bounding_box': '76.08166,-67.1746,88.19689,21.04862',
                            'page_size': 200,
                            },
                        headers={
                            'Accept': 'application/json'
                            }
                       )
print(response.status_code)
# confine to bounding box, time range, and certain dataset
#C2036878004-POCLOUD
#C1990404821-POCLOUD
```

```python
print(response.headers['CMR-Hits'])
```

```python
granules = response.json()['feed']['entry']
for granule in granules:
    print(f'{granule["dataset_id"]} {granule["id"]}')
```
